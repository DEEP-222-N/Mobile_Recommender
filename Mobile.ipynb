```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import NearestNeighbors
import pickle
```

```python
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Projects/ML/cleaned_mobile_dataset.csv")
```

```python
df.head()
```

```python
features = ['Launched Price (INR)', 'RAM_GB', 'Storage_GB', 'Battery Capacity(mAh)','Back Cam Primary (MP)', 'Front Cam Primary (MP)']
```

```python
df_filtered = df.dropna(subset=features)
```

```python
df_filtered.isna().sum()
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_filtered[features])
```

```python
knn = NearestNeighbors(n_neighbors=3, metric='euclidean')
knn.fit(X_scaled)
```

```python
# Save model and data
with open("knn_model.pkl", "wb") as f:
    pickle.dump((knn, scaler, df_filtered), f)
```
